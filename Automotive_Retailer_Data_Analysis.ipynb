# **Year Of Registration**

```python
data.loc[ data['year_of_registration'].isnull()].head()
```

This header insists that the missing values in 'year of registration are caused by the car having yet to be registered.

```python
data['year_of_registration'].describe()
```

Since the newest registered cars are registered as 2020, I'll fill all new cars with the year of registration as 2020.

```python
data['year_of_registration'] = data['year_of_registration'].fillna('2020')
data['year_of_registration'] = pandas.to_numeric(data['year_of_registration'])
```

The description also says that the oldest car was registered in 999 which must be an anomaly.

```python
data.nsmallest(30, 'year_of_registration')
```

Looking at the oldest cars, it seems like a common error and so i cannot just remove the anomalous result, as there may be harder to notice, but similar, errors throughout the dataframe. I will use the Registration Code of the car to verify it's age, where possible.

```python
mask = data['reg_code'].astype(str).str.isdigit();
filteredRegCodes = data[mask];
filteredRegCodes['reg_code'] = pandas.to_numeric(filteredRegCodes['reg_code']);
filteredRegCodes['year_of_registration'] = filteredRegCodes['year_of_registration'].astype(int);
```

Here I am removing any string values from the column 'reg_codes' so that i can compare the registration codes to the year of registration. This will allow me to verify the age of the vast majority of the cars in the dataframe.

```python
for index, row in filteredRegCodes.iterrows():

        if ((row['year_of_registration'] - row['reg_code']) != 1950 or (row['year_of_registration'] - row['reg_code']) != 1951  or (row['year_of_registration'] - row['reg_code']) != 2000):
          if row['reg_code'] > 50:
            if (1950 + row['reg_code']) < 2020 and (1950 + row['reg_code']) >= 1973:
             data.loc[index, 'year_of_registration'] = (1950 + row['reg_code']);
          else:
           if (2000 + row['reg_code']) <= 2020:
            data.loc[index, 'year_of_registration'] = (2000 + row['reg_code']);
```

Checks that the reg plate holds up to the modern system, and if not, calculates what the correct year of registration should be. If this year happens to fall within the range of the modern system, the year of registration in the main dataframe is replaced.

I Shouldn't need to add the reg_code for all of the new cars, as it will not be used for price comparisons later on. In cases where the reg code was outdated, or a string, it was ignored as the reg_code will not be used and so it won't skew the data.

# **Fuel Type**

```python
fueldata = data['fuel_type'].value_counts()
fueldata
```

```python
print(len(data.loc[ data['fuel_type'].isnull()]))
```

This Column also contains multiple Null results, although i believe this one will be easier to fix. I will, where possible, find a car of an identical model and replace the null fuel type with that of the similar car.

```python
nullFuel = data.loc[ data['fuel_type'].isnull()]
```

```python
for index, row in nullFuel.iterrows():
  identicalModel = data[data['standard_model'] == row['standard_model']].index[0]
  data.loc[index, 'fuel_type'] = data.loc[identicalModel,'fuel_type']
```

# **Feature Engineering**

The First thing i will be adding to the database is a 'Price Brackets' column. This will allow easy use of data exclusively within a particular quartile of Price, important when looking at the concentration of catagorical values at different price points.

```python
data['price'].describe()
```

```python
for index, row in data.iterrows():

        if (row['price'] >= 20000):
          data.loc[index, 'price_bracket'] = 4;
        else:
         if (row['price'] >= 12600):
          data.loc[index, 'price_bracket'] = 3;
         else:
          if (row['price'] >= 7495):
            data.loc[index, 'price_bracket'] = 2;
          else:
            data.loc[index, 'price_bracket'] = 1;
```

```python
data.head()
```
